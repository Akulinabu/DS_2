- простые графики

```python
# имитация данных
#  массив (вектор-строка) из 1000 значений
# линейной последовательности от 0 до 10
x=np.linspace(0, 10, 1000)
x
```

```python
# данные для оси у
y=np.sin(x)
```

```python
# рисуем график
plt.plot(x, y)
plt.show()
```

```python
# совмещение двух графиковв одном окне
plt.plot(x, y)
plt.plot(x, np.cos(x))
plt.show()
```

```python
# интерактивный режим
%matplotlib notebook

# метки и легенды графиков

# легенды
plt.plot(x, y, label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()
# заголовок
plt.title("Синус и косинус от х")

# метки
plt.xlabel("x")
plt.ylabel("sin(x) & cos(x)")

plt.show()
```

```python
# цвет и вид линий

# цвета можно задавать разными способами
plt.plot(x, np.sin(x), color='blue') # по названию
plt.plot(x, np.sin(x+1), color='g') # по перечню букв цветов (rgbcmyk)
plt.plot(x, np.sin(x+2), color='0.75') # оттенки серого в диапазоне от 0 до 1
plt.plot(x, np.sin(x+3), color=(1.0, 0.3, 0.2)) #B-кортеж, значения от 0 до 1
```

```python
# вид линий
plt.plot(x, x+3, linestyle='--') # штриховая линия
plt.plot(x, x+2, linestyle=':') # пунктирная линия
plt.plot(x, x+1, linestyle='-.') # штрих-пунктирная линия
plt.show()
```

```python
# объединение кодов вида
plt.plot(x, x+3, '--g') # штриховая линия
plt.plot(x, x+2, ':k') # пунктирная линия
plt.plot(x, x+1, '-.r') # штрих-пунктирная линия
plt.show()
```
